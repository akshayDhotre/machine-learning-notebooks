```python
# Import required libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Import dataset and create feature and result
dataset = pd.read_csv('datasets/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
# Train random forest regressor model
from sklearn.ensemble import RandomForestClassifier
rfr = RandomForestClassifier(n_estimators = 500, random_state = 0)
rfr.fit(X, y)
```

```python
# Predict the salary for given position
rfr.predict([[6.5]])
```

```python
# visualize random forest classifier values for given data
X_grid = np.arange(min(X), max(X), 0.001)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y)
plt.plot(X_grid, rfr.predict(X_grid), color='red')
```
