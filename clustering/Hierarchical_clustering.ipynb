```python
# Import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import dataset
dataset = pd.read_csv('../datasets/Mall_Customers.csv')
```

```python
# remove unwanted features like customer ID, ID
# Also for proper visualization choosing only 2 features for K means clustering
X = dataset.iloc[:, [3, 4]].values
```

```python
# Using dendrogram to choose optimal number of clusters
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidian Distance')
plt.show()
```

```python
# Train hierarchical clusterer model for optimal clusters
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
# Visualize the clusters
plt.scatter(X[y_hc==0, 0], X[y_hc==0, 1], color='red', label='Cluster 1')
plt.scatter(X[y_hc==1, 0], X[y_hc==1, 1], color='blue', label='Cluster 2')
plt.scatter(X[y_hc==2, 0], X[y_hc==2, 1], color='cyan', label='Cluster 3')
plt.scatter(X[y_hc==3, 0], X[y_hc==3, 1], color='green', label='Cluster 4')
plt.scatter(X[y_hc==4, 0], X[y_hc==4, 1], color='magenta', label='Cluster 5')
plt.title("Clusters")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
